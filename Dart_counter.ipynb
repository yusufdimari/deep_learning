```python
def dart_game():
    print("ðŸŽ¯ Welcome to the Dart Game!")
    print("Each player starts with 201 points.")
    print("Each player gets 3 throws per turn.")
    print("If you score more than your remaining points in a throw, that throw doesn't count!\n")

    players = ["Aliyu", "Yusuf"]
    scores = {player: 101 for player in players}
    turn = 0

    while True:
        current_player = players[turn % 2]
        print(f"\n{current_player}'s turn. Current score: {scores[current_player]}")

        for throw_num in range(1, 4):
            print(f"Throw {throw_num}/3")
            try:
                hit = int(input("Enter score for this throw: "))
            except ValueError:
                print("Invalid input. Please enter a number.")
                continue

            if hit < 0:
                print("Score can't be negative. Try again.")
                continue

            if hit > scores[current_player]:
                print(f"Too high! You hit {hit}, but only {scores[current_player]} remaining. No score counted.")
            else:
                scores[current_player] -= hit

            if scores[current_player] == 0:
                print(f"\nðŸŽ¯ {current_player} wins the game!")
                return

        turn += 1

# Run the game
dart_game()
```
